```python
# Drop unnessory Column
df = df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
df.duplicated().sum()
```

```python
# Add Attrition Column
df['Attrition_Code'] = df['Attrition'].map({'Yes':1, 'No':0})

print(df[['Attrition', 'Attrition_Code']].head())
```

# Attrition Count

```python
# Check Attrition distribution
print(df['Attrition'].value_counts())
ax = sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title('Attrition Distribution', fontsize=14, fontweight='bold')

#data label
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 2, 
            int(height), ha='center', color='black')
plt.show()
```

# Histogram: Attrition by Age

```python
# Attrition vs Age
plt.figure(figsize=(12,6))
ax1 = sns.histplot(data=df, x='Age', hue='Attrition', kde=True, palette='Set1')
plt.title('Attrition by Age', fontsize=14, fontweight='bold')

#data label
for p in ax1.patches:
    height = p.get_height()
    if height > 0:
        ax1.text(p.get_x() + p.get_width()/2., height + 2, 
            int(height), ha='center', fontsize=9,  color='black')
    
plt.show()
```

```python
sns.set(style='whitegrid', palette='pastel', font_scale=1.1)
```

# Attrition by Department

```python
# Attrition vs Department + JobRole (Stacked Countplot)
plt.figure(figsize=(14,6))
ax2 = sns.countplot(x='Department', hue='Attrition', data=df, palette='plasma')
plt.title('Attrition by Department', fontsize=14, fontweight='bold')
#data label
for p in ax2.patches:
    height = p.get_height()
    if height > 0:
         ax2.text(p.get_x() + p.get_width()/2., height + 2, 
            int(height), ha='center', color='black')
plt.show()
```

# Violinplot: Attrition by MonthlyIncome

```python
# MonthlyIncome vs Attrition (violine plot)
plt.figure(figsize=(12,6))
sns.violinplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set3')
plt.title('Monthly Income vs Attrition', fontsize=14, fontweight='bold')
plt.show()
```

# Scatter Plot: Age & JobSatisfaction by Attrition

```python
# Age vs Job Satisfaction vs Attrition
plt.figure(figsize=(12,6))
sns.scatterplot(x='Age', y='JobSatisfaction', hue='Attrition', data=df, palette='Set1', alpha=0.7, s=100)
plt.title('Age vs JobSatisfaction vs Atrition', fontsize=14, fontweight='bold')
plt.show()
```

# Distance From Home by Attrition

```python
# DistanceFromHome vs Attrition (Boxplot)
distance_counts = df.groupby(['Attrition', 'DistanceFromHome']).size().reset_index(name='counts')
plt.figure(figsize=(12,6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df, palette='Set3')
plt.title('Distance From Home vs Attrition', fontsize=14, fontweight='bold')
plt.show()
```

# Overtime by Attrition Percentage

```python
# OverTime vs Attrition (Percentage Stacked)
overtime_attrition = df.groupby(['OverTime', 'Attrition']).size().unstack(fill_value=0)
overtime_attrition_pct = overtime_attrition.div(overtime_attrition.sum(axis=1), axis=0)*100
overtime_attrition_pct.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Accent')

#data label
for i, overtime in enumerate(overtime_attrition_pct.index):
    bottom=0
    for j, attr in enumerate(overtime_attrition_pct.columns):
        perc = overtime_attrition_pct.loc[overtime, attr]
        plt.text(i, bottom + perc/2, f"{perc:.1f}%", ha='center', va='center', color='white', fontsize=10)
        bottom+= perc
plt.ylabel('Percentage')
plt.title('Attrition vs OverTime', fontsize=14, fontweight='bold')
plt.show()
```

# Education by Attrition

```python
#Education vs Attrition (Countplot)
plt.figure(figsize=(12,6))
ax3 = sns.countplot(x='EducationField', hue='Attrition', data=df, palette='Set2')

#data label
for p in ax3.patches:
    height = p.get_height()
    if height > 0:
         ax3.text(p.get_x() + p.get_width()/2., height + 2, 
            int(height), ha='center', color='black')
plt.title('Education Field vs Attrition', fontsize=14, fontweight='bold')
plt.show()
```

# Heatmap: WorkLifeBalance by Attrition

```python
# WorkLifeBalance vs Attrition (Heatmap)
wlb_counts = df.pivot_table(index='WorkLifeBalance', columns='Attrition', aggfunc='size', fill_value=0)
plt.figure(figsize=(10,6))
sns.heatmap(wlb_counts, annot=True, cmap="plasma", fmt='d', linewidths=0.5)
plt.title('WorkLifeBalance vs Attrition Rate', fontsize=14, fontweight='bold')
plt.show()
```

# EnvironmentSatisfaction by JobSatisfaction by Attrition

```python
# EnvironmentSatisfaction vs JobSatisfaction vs Attrition
g = sns.FacetGrid(df, col='Attrition', height=5, aspect=1)
g.map_dataframe(sns.scatterplot, x='EnvironmentSatisfaction', y='JobSatisfaction', hue='Attrition', palette='Set2', alpha=0.7, s=80)
g.set_axis_labels('Environment Satisfaction', 'Job Satisfaction')
g.set_titles('Attrition = {col_name}')
g.add_legend()
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Environment vs Job Satisfaction by Attrition', fontsize=14, fontweight='bold')
plt.show()
```

# Avg Department wise by Attrition 

```python
# Department-wise Attrition Rate
dept_attrition = df.groupby('Department')['Attrition_Code'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=dept_attrition.index,y=dept_attrition.values, palette='Set3')
plt.title('Department-wise Avg Attrition Rate', fontsize=14, fontweight='bold')
plt.ylabel('Attrition Rate')
plt.ylim(0,1)
plt.show()
```

# KDE Plot: YearsAtComapny bt Attrition

```python
plt.figure(figsize=(12,6))
sns.kdeplot(data=df[df['Attrition_Code']==1]['YearsAtCompany'], label='Left', fill=True)
sns.kdeplot(data=df[df['Attrition_Code']==0]['YearsAtCompany'], label='Stayed', fill=True)
plt.title('Years at Company vs Attrition Rate', fontsize=14, fontweight='bold')
plt.xlabel('YearsAtCompany')
plt.legend()
plt.show()
```

# Job Role by Avg Attrition

```python
#JObRole-wise Attrition Rate
role_attrition = df.groupby('JobRole')['Attrition_Code'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12,6))
sns.barplot(x='JobRole',y='Attrition_Code', data=role_attrition, palette='Set3')
#Label on bar
for i, row in role_attrition.iterrows():
    plt.text(i, row['Attrition_Code'] + 0.01, f"{row['Attrition_Code']:.2f}", ha='center',fontsize=10)
    

plt.title('JobRole-wise Attrition Rate', fontsize=14, fontweight='bold')
plt.xlabel('Job Role')
plt.ylabel('Mean Attrition Rate')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.show()
role_attrition
```

# Worklifebalance vs Department by Attrition

# Attrition by Gender vs Department

```python
# Attrition vs Gender vs Department (FacetGrid)
g = sns.FacetGrid(df, col='Department', height=4, col_wrap=3)
g.map_dataframe(sns.countplot, x='Gender', hue='Attrition', palette='Set2', dodge=True)

# data label
for ax in g.axes.flat:
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.text(p.get_x() + p.get_width()/2., height + 0.1,
                   int(height), ha='center', va='bottom', fontsize=9)
            
g.add_legend()
g.fig.suptitle('Gender-wise Attrition by Department', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```

# Heatmap: Environment Satisfaction by Job Role

```python
# EnvironmentSatisfaction vs JobRole (Heatmap)
env_role = df.pivot_table(index='JobRole', columns='EnvironmentSatisfaction', values='Attrition_Code', aggfunc='mean')
plt.figure(figsize=(12,6))
sns.heatmap(env_role, annot=True, fmt='.2f', cmap='plasma')
plt.title('Attrition by JobRole & EnvironmentSatisfaction', fontsize=14, fontweight='bold')
plt.show()
```

# OverTime vs Department 

```python
# Overtime vs Deapartment (Stacked Percentage)
ot_dept = pd.crosstab(df['Department'], df['OverTime'])
ot_dept_pct = ot_dept.div(ot_dept.sum(axis=1), axis=0)*100
ot_dept_pct.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Accent')
#data label
for i, dept in enumerate(ot_dept_pct.index):
    bottom=0
    for j, attr in enumerate(ot_dept_pct.columns):
        perc = ot_dept_pct.loc[dept, attr]
        if perc > 0:
            plt.text(i, bottom + perc/2, f"{perc:.1f}%", ha='center', va='center', color='white', fontsize=10)
        bottom+= perc
plt.title('OverTime Distribution by Department', fontsize=14, fontweight='bold')
plt.ylabel('Percentage of Employees')
plt.show()
```

# Attrition Risk Score (Department + JOb Role) - Combine KPI Bar

```python
# Attrition Risk Score (Department + JobRole) - Combined KPI Bar
risk_score = df.groupby(['Department', 'JobRole'])['Attrition_Code'].mean().sort_values(ascending=False).reset_index()
risk_score.plot(kind='bar', figsize=(12,6), color='tomato')
#Label on bar
for i, row in risk_score.iterrows():
    plt.text(i, row['Attrition_Code'] + 0.01, f"{row['Attrition_Code']:.2f}", ha='center',fontsize=10)
    

plt.title('High Risk Roles by Department', fontsize=14, fontweight='bold')
plt.ylabel('Attrition Rate')
plt.show()
```

# KPI Calculations: Total Employee, Leave Employee, Avg Monthly Income, Avg Age Employee

```python
TotalEmployees = df['EmployeeNumber'].value_counts().sum()
LeavedEmployee = df['Attrition_Code'].sum()
AverageMonthlyIncome = df['MonthlyIncome'].mean().round(2)
AvgrageAgeEmp = df['Age'].mean().round(2)

kpis_title = ['TotalEmployees', 'LeavedEmployee', 'AverageMonthlyIncome', 'AvgrageAgeEmp']
kpis = [
    f"{Employees:,.0f}",
    f"{EmployeesLeave:,}",
    f"{AverageMonthlyIncome}",
    f"{AvgrageAgeEmp}"]
sns.set_style('white')
fig, axes = plt.subplots(1, 4, figsize=(20,3))
colors=['#2ecc71','#3498db','#e67e22','#9b59b6']

for ax, title, value, color in zip(axes, kpis_title, kpis, colors):
    ax.set_facecolor(color)
    ax.text(0.5, 0.55, value, ha='center', va='center', fontsize=20, color='white', fontweight='bold')
    ax.text(0.5, 0.25, title, ha='center', va='center', fontsize=20, color='white', fontweight='bold')
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

plt.suptitle('KPI', fontsize=18, fontweight='bold')
plt.subplots_adjust(top=0.8, wspace=0.3)
plt.show()
```
